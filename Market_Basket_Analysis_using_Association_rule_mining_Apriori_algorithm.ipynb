```python
#Importing necessary Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori

#Loading the dataset
df = pd.read_csv("//content/Groceries_dataset.csv")

#Viewing first few rows of the dataset
print(df.head(10))
```

```python
df.describe()
```

```python
#Data Cleaning
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df['Member_number'].nunique()
```

```python
#Grouping the rows by customers
df_grouped = df.groupby(['Member_number','Date'])['itemDescription'].apply(sum)
df_grouped
```

```python
#Creating the list of transactions
transactions = [a[1]['itemDescription'].tolist() for a in list(df.groupby(['Member_number','Date']))]
transactions
```

```python
#Encoding the transactions
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
te_ary
```

```python
transactions = pd.DataFrame(te_ary, columns=te.columns_)
pf = transactions.describe()
pf
```

```python
pf.iloc[0]-pf.iloc[3]
```

```python
f = pf.iloc[0]-pf.iloc[3]
a = f.tolist()
b = list(f.index)
item = pd.DataFrame([[a[r],b[r]] for r in range(len(a))], columns = ['Count','Item'])
item = item.sort_values(['Count'], ascending = False).head(50)
transactions
```

**Apriori Algorithm**

```python
#Computing frequent itemsets
frequent_itemsets = apriori(transactions, min_support = 0.0005, max_len = 4, use_colnames = True)

#Printing number of itemsets
print(len(frequent_itemsets))
```

```python
print(frequent_itemsets.head())
```

**Apriori and Computing Association Rule**

```python
#Computing association rules
Arules = association_rules(frequent_itemsets,
                           metric = "support",
                           min_threshold = 0.001)
```

```python
Arules
```

```python
#Raising the threshold
#Computing association rules
Arules_2 = association_rules(frequent_itemsets,
                             metric = "support",
                             min_threshold = 0.010)

Arules_2
```

```python
#Raising the threshold
#Computing association rules
Arules_3 = association_rules(frequent_itemsets,
                             metric = "support",
                             min_threshold = 0.050)
Arules_3
```

```python
filtered_rules = Arules[(Arules['antecedent support']>0.06) & 
                        (Arules['support']>0.002)&
                        (Arules['confidence']>0.04)&
                        (Arules['lift']>1.00)]
filtered_rules
```
